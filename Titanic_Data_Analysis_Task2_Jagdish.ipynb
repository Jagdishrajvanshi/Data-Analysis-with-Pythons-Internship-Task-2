```python
# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load Titanic Dataset
df = sns.load_dataset("titanic")

# Step 3: Preview Dataset
print("Dataset Preview:")
display(df.head())

# Step 4: Dataset Info
print("\nDataset Information:")
df.info()

# Step 5: Check Missing Values
print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

# Step 6: Data Cleaning
# Fill missing age values with median
df['age'] = df['age'].fillna(df['age'].median())

# Drop rows with missing embarked values
df = df.dropna(subset=['embarked'])

# Step 7: Verify Missing Values After Cleaning
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# Step 8: Survival Rate by Gender
plt.figure()
sns.barplot(x='sex', y='survived', data=df)
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()

# Step 9: Survival Rate by Passenger Class
plt.figure()
sns.barplot(x='pclass', y='survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()

# Step 10: Age Distribution Histogram
plt.figure()
plt.hist(df['age'], bins=20)
plt.title("Passenger Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()

# Step 11: Survival Analysis Tables
print("\nSurvival Rate by Gender:")
print(df.groupby('sex')['survived'].mean())

print("\nSurvival Rate by Passenger Class:")
print(df.groupby('pclass')['survived'].mean())

# Step 12: Final Conclusion
print("\nConclusion:")
print("""
Females had a higher survival rate compared to males.
Passengers traveling in higher classes had better chances of survival.
Age distribution shows most passengers were between 20 and 40 years old.
Gender and passenger class significantly influenced survival.
""")
```
